```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

if __name__ == '__main__':
    data = pd.read_csv("/home/pks/Downloads/haberman.csv")
    
#     print(data)
    data.boxplot(column=['age', 'operation_year', 'axil_nodes'], by='status')
    
    plt.show()
```

```python
%matplotlib inline

if __name__ == '__main__':
    plt.scatter(data[data.status==1]['age'], data[data.status==1]['axil_nodes'], s=14, color='red')
    plt.scatter(data[data.status==2]['age'], data[data.status==2]['axil_nodes'], s=14, color='green')
    plt.xlabel('age')
    plt.ylabel('axil_nodes')
    plt.show()
    
    plt.scatter(data[data.status==1]['age'], data[data.status==1]['operation_year'], s=8, color='red')
    plt.scatter(data[data.status==2]['age'], data[data.status==2]['operation_year'], s=8, color='green')
    plt.xlabel('age')
    plt.ylabel('operation_year')
    plt.show()
    
    plt.scatter(data[data.status==1]['axil_nodes'], data[data.status==1]['operation_year'], s=8, color='red')
    plt.scatter(data[data.status==2]['axil_nodes'], data[data.status==2]['operation_year'], s=8, color='green')
    plt.xlabel('axil_nodes')
    plt.ylabel('operation_year')
    plt.show()
```
